## Άσκηση 19

Αρχικοποιούμε την αλυσίδα

```python
import random
import numpy as np
random.seed(2018)
from simple_markov_chain_lib import markov_chain

p = 1/6

init_probs = {1: 1.0} 

# Stochastic matrix
markov_table = {
    1: {2: 0.5, 3: 0.5},
    2: {1: 1/3, 2: 1/3, 3: 1/3},
    3: {1: p, 2: 2/3-p, 3: 1/3}
}
 
mc = markov_chain(markov_table, init_probs)
```

Συγκεκριμένα, θα εκτιμήσουμε την πιθανότητα $\mathbb{P}_1[X_{20}= 3]=\mathbb{P}\left[X_{20} = 3 \mid X_0 = 1 \right]$. 
Για να το κάνουμε αυτό, θα τρέξουμε την αλυσίδα πολλές φορές (παράμετρος `N`) ξεκινώντας πάντα από την κατάσταση `1` και θα μετρήσουμε σε πόσες από αυτές η αλυσίδα βρέθηκε στην κατάσταση `3` μετά από `steps` βήματα.

Από τον ΝΜΑ γνωρίζουμε ότι

$$ \lim_{N\to\infty}\frac{Y_1+\cdots+Y_N}{N}=\mathbb{E}_1[Y]=\mathbb{P}_1[X_{20}=3].$$

Άρα η εκτίμηση που θα κάνουμε θα πρέπει να έχει τέτοιο Ν ώστε να μας δίνει ικανοποιητικά αποτελέματα που συμφωνούν με τη θεωρία.


```python
def experiment(mc, N = 200, steps = 20, state = 3, runs = 50):
    estimates = []
    for r in range(runs):
        counter = 0
    
        for i in range(N):
            mc.start()  
            for j in range(steps):  mc.move()
            if mc.running_state == state:  counter += 1

        phat = counter / N
        print ("Run number: {}".format(r+1))
        print(
            """
            We executed {0} times the first {1} steps of the markov chain
            and we captured the running state in state one {2} times.
            So we estimate the Pr[X_{1} = {4} | X_0 = 1] to be {3}
            """.format(N, steps, counter, phat, state)
        )
        estimates.append(phat)
    return np.array(estimates)
        

estimates = experiment(mc);

print ("""Statistical analysis for 50 runs: 
        Mean of estimates : {0:.5f}
        Standard deviation : {1:.5f} (unbiased)""".format(np.mean(estimates), np.std(estimates, ddof = 1)))
```

```python
estimates2 = experiment(mc, N = 20000);

print ("""Statistical analysis for 50 runs: 
        Mean of estimates : {0:.5f}
        Standard deviation : {1:.5f} (unbiased)""".format(np.mean(estimates2), np.std(estimates2, ddof = 1)))
```

Παρατηρούμε ότι τα αποτελέσματά μας **συμφωνούν με τα θεωρητικά**.

## Άσκηση 40

Το παιχνίδι τένις περιγράφεται από τον ακόλουθο γράφο καταστάσεων

![Markov Chain of Tennis Game](https://www.wolfram.com/mathematica/new-in-9/markov-chains-and-queues/HTMLImages.en/analyze-a-tennis-game/O_23.png)

Ο πρώτος παίκτης σκοράρει με πιθανότητα $p$ και ο δεύτερος με $q = 1- p$

Ο πίνακας μεταβάσεων και η αλυσίδα είναι οι ακόλουθοι

```python
p = 0.6
init_probs_tennis = {'0-0' : 1.}
matrix = {
    '0-0' : {'15-0' : p, '0-15' : 1 - p},
    '15-0' : {'30-0': p, '15-15' : 1 - p},
    '0-15' : {'15-15': p, '0-30' : 1 - p},
    '30-0' : {'40-0' : p, '30-15' : 1 - p},
    '15-15' : {'30-15' : p, '15-30' : 1-p},
    '0-30' : {'15-30' : p, '0-40' : 1-p},
    '40-0' : {'40-15' : 1 - p, 'GameA' : p},
    '30-15' : {'40-15' : p, 'Deuce' : 1 - p},
    '15-30' : {'Deuce' : p, '15-40' : 1-p},
    '0-40' : {'15-40' : p, 'GameB': 1 - p},
    '40-15' : { 'GameA' : p, 'AdvA' : 1 - p},
    '15-40' : {'GameB' : 1 - p, 'AdvB' : p},
    'GameA' : {'GameA' : 1.0},
    'GameB' : {'GameB' : 1.0},
    'AdvA' : {'GameA' : p, 'Deuce' : 1-p},
    'AdvB' : {'GameB' : 1-p, 'Deuce' : p},
    'Deuce' : {'AdvA' : p, 'AdvB': 1-p}
}

tennis = markov_chain(matrix, init_probs_tennis)
```

Θέλουμε την πιθανότητα $\mathbb P [X_s = \text{"GameA"} \mid X_0 = \text{"0-0"} ]$.

Θα βάλουμε την αλυσίδα να τρέξει `s = 20` steps για `N = 2000` και θα τρέξουμε το συνολικό πείραμα 50 φορές

```python
estimates_tennis = experiment(tennis, N = 2000, steps = 20, state = 'GameA', runs = 50)
print ("""Statistical analysis for 50 runs: 
        Mean of estimates (Probability of winning the match) : {0:.5f}
        Standard deviation (error) : {1:.5f} (unbiased)""".format(np.mean(estimates_tennis), np.std(estimates_tennis, ddof = 1)))
```

Παρατηρούμε ότι τα επιμέρους πειράματα δίνουν σχεδόν το ίδιο αποτέλεσμα με πολύ μικρό σφάλμα (τάξης χιλιοστού).
